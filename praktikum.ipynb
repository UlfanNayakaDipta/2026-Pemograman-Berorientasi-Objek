## Praktikum 3

```python
class Buku:
    def __init__(self, judul, pengarang, tahun_terbit):
        self.judul = judul
        self.pengarang = pengarang
        self.tahun_terbit = tahun_terbit
        self.status = "Tersedia"

    def tampilkan_info(self):
        print(f"Judul: {self.judul}")
        print(f"Pengarang: {self.pengarang}")
        print(f"Tahun Terbit: {self.tahun_terbit}")
        print(f"Status: {self.status}")

    def pinjam(self):
        if self.status == "Tersedia":
            self.status = "Dipinjam"
            print(f"Buku '{self.judul}' telah dipinjam.")
        else:
            print(f"Buku '{self.judul}' sedang dipinjam.")

    def kembalikan(self):
        if self.status == "Dipinjam":
            self.status = "Tersedia"
            print(f"Buku '{self.judul}' telah dikembalikan.")
        else:
            print(f"Buku '{self.judul}' tidak sedang dipinjam.")


buku1 = Buku("Pemrograman Python", "John Doe", 2021)
buku2 = Buku("Data Science untuk Pemula", "Jane Smith", 2020)

buku1.tampilkan_info()
buku2.pinjam()
buku2.kembalikan()
buku1.pinjam()
buku1.tampilkan_info()


# Class Mahasiswa
class Mahasiswa:
    def __init__(self, nama, nim, umur):
        self.nama = nama
        self.nim = nim
        self.umur = umur

    def tampilkan_info(self):
        print(f"Nama: {self.nama}")
        print(f"NIM: {self.nim}")
        print(f"Umur: {self.umur} tahun")

    def ubah_umur(self, umur_baru):
        self.umur = umur_baru


mahasiswa1 = Mahasiswa("Andi", "12345", 20)
mahasiswa2 = Mahasiswa("Budi", "67890", 22)

mahasiswa1.tampilkan_info()
print()
mahasiswa2.tampilkan_info()

mahasiswa1.ubah_umur(21)

print("\nSetelah diubah:")
mahasiswa1.tampilkan_info()
```

## Praktikum 4

```python
class Mobil:
    def __init__(self, merk, warna, tahun, harga):
        self.merk = merk
        self.warna = warna
        self.tahun = tahun
        self.harga = harga

    def tampilkan_info(self):
        print(f"Mobil {self.merk}, warna {self.warna}, tahun {self.tahun}, harga {self.harga}")

    def diskon(self, persen):
        diskon = self.harga * (persen / 100)
        harga_akhir = self.harga - diskon
        print(f"Harga setelah diskon {persen}%: {harga_akhir}")

    def hitung_usia(self, tahun_sekarang):
        return tahun_sekarang - self.tahun

    def update(self, harga_baru, tahun_baru):
        self.harga = harga_baru
        self.tahun = tahun_baru
        print("Data mobil diperbarui")


mobil1 = Mobil("Toyota", "Hitam", 2015, 300000000)
mobil2 = Mobil("Honda", "Merah", 2018, 250000000)

mobil1.tampilkan_info()
mobil2.tampilkan_info()

mobil1.diskon(10)
mobil2.diskon(15)

print("Usia mobil1:", mobil1.hitung_usia(2025))

mobil1.update(280000000, 2016)
mobil1.tampilkan_info()
```

## Program

```python
# Program Penjualan HP dengan Menu, Diskon, dan Update Stok
class HP:
    def __init__(self, nama, harga, stok):
        self.nama = nama
        self.harga = harga
        self.stok = stok

    def tampilkan(self):
        print(f"{self.nama}")
        print(f"Harga : Rp {self.harga:,}")
        print(f"Stok  : {self.stok}")
        print("-" * 30)

    def beli(self, jumlah):
        if jumlah <= self.stok:
            self.stok -= jumlah
            total = jumlah * self.harga
            return total
        else:
            print("❌ Stok tidak mencukupi!")
            return None

# fungsi tambahan (modular)
def hitung_diskon(total):
    if total >= 20000000:
        diskon = 0.1 * total
    elif total >= 10000000:
        diskon = 0.05 * total
    else:
        diskon = 0
    return diskon
```

```python
# Data produk
hp1 = HP("Samsung Galaxy S26 Ultra", 22000000, 5)
hp2 = HP("Xiaomi 17 Ultra", 18000000, 7)

while True:
    print("\n=== TOKO HP ===")
    print("1. Lihat Produk")
    print("2. Beli HP")
    print("3. Keluar")
    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        print("\nDaftar Produk:")
        hp1.tampilkan()
        hp2.tampilkan()

    elif pilihan == "2":
        print("\nPilih Produk:")
        print("1. Samsung Galaxy S26 Ultra")
        print("2. Xiaomi 17 Ultra")
        pilih_hp = input("Masukkan pilihan: ")
        jumlah = int(input("Jumlah beli: "))

        if pilih_hp == "1":
            total = hp1.beli(jumlah)
            nama_hp = hp1.nama
        elif pilih_hp == "2":
            total = hp2.beli(jumlah)
            nama_hp = hp2.nama
        else:
            print("❌ Pilihan tidak valid")
            continue

        if total is not None:
            diskon = hitung_diskon(total)
            total_bayar = total - diskon

            print("\n=== STRUK PEMBELIAN ===")
            print("Produk :", nama_hp)
            print("Total  : Rp", format(total, ","))
            print("Diskon : Rp", format(int(diskon), ","))
            print("Bayar  : Rp", format(int(total_bayar), ","))
            print("========================")

    elif pilihan == "3":
        print("Terima kasih!")
        break

    else:
        print("❌ Menu tidak tersedia")
```
