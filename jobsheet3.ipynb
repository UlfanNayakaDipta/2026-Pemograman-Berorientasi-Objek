## Praktikum 1

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.__owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"{amount} telah ditambahkan ke akun {self.__owner}.")
        else:
            print("Jumlah deposit harus lebih dari 0.")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"{amount} telah ditarik dari akun {self.__owner}.")
        else:
            print("Saldo tidak mencukupi.")

    def get_balance(self):
        return self.__balance


# Main
if __name__ == "__main__":
    alice_account = BankAccount("Alice", 1000)

    alice_account.deposit(500)
    alice_account.deposit(-100)

    alice_account.withdraw(300)
    alice_account.withdraw(2000)

    print(f"Saldo akhir: {alice_account.get_balance()}")
```

## Praktikum 2

```python
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    def update_salary(self, increase):
        if increase > 0:
            self.__salary += increase
            print(f"Gaji naik sebesar {increase}")
        else:
            print("Kenaikan harus > 0")

    def set_salary(self, new_salary):
        if new_salary >= 0:
            self.__salary = new_salary
            print(f"Gaji diubah menjadi {new_salary}")
        else:
            print("Gaji tidak boleh negatif")

    def get_salary(self):
        return self.__salary

    def get_employee_info(self):
        return f"Nama: {self.__name}, Gaji: {self.__salary}"


# Main
if __name__ == "__main__":
    emp = Employee("John Doe", 50000)

    print(emp.get_employee_info())

    emp.update_salary(5000)
    print(emp.get_salary())

    emp.set_salary(60000)
    print(emp.get_employee_info())
```

## Praktikum 3

```python
class SimpleExample:
    def __init__(self, name):
        self.name = name
        print(f"Konstruktor: Objek '{self.name}' dibuat")

    def __del__(self):
        print(f"Destruktor: Objek '{self.name}' dihapus")


def main():
    print("Program dimulai\n")

    obj = SimpleExample("Demo")

    print("Program berjalan\n")

    del obj

    print("Objek sudah dihapus\n")
    print("Program selesai")


if __name__ == "__main__":
    main()
```

## Praktikum 4

```python
class FileLogger:
    def __init__(self, filename):
        self.filename = filename
        try:
            self.file = open(filename, "a")
            print(f"File {filename} dibuka")
        except Exception as e:
            print(f"Gagal buka file: {e}")

    def write_log(self, message):
        self.file.write(message + "\n")
        self.file.flush()
        print(f"Log: {message}")

    def __del__(self):
        if hasattr(self, "file") and not self.file.closed:
            self.file.close()
            print(f"File {self.filename} ditutup")


# Main
if __name__ == "__main__":
    logger = FileLogger("app.log")

    logger.write_log("Aplikasi mulai")
    logger.write_log("Proses jalan")
    logger.write_log("Selesai")

    del logger
```

## Praktikum 5

```python
class Calculator:
    def __init__(self, initial_value=0):
        self.value = initial_value
        print(f"Nilai awal: {self.value}")

    def add(self, number):
        self.value += number
        print(f"Tambah {number} → {self.value}")

    def subtract(self, number):
        self.value -= number
        print(f"Kurang {number} → {self.value}")

    def reset(self):
        self.value = 0
        print("Reset ke 0")

    def show_value(self):
        print(f"Nilai sekarang: {self.value}")


# Main
if __name__ == "__main__":
    calc1 = Calculator(10)
    calc1.add(5)
    calc1.subtract(3)
    calc1.show_value()

    calc2 = Calculator()
    calc2.add(20)
    calc2.show_value()
```

## Praktikum 6

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        if value:
            self.__name = value
        else:
            print("Nama tidak boleh kosong")

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if value >= 0:
            self.__age = value
        else:
            print("Umur tidak boleh negatif")


# Main
if __name__ == "__main__":
    p = Person("Alice", 30)

    print(p.name, p.age)

    p.name = "Bob"
    p.age = 35

    print(p.name, p.age)

    p.age = -5
```

## Program 1

```python
class Mahasiswa:
    def __init__(self, nama, nim):
        self.__nama = nama
        self.__nim = nim
        print(f"Objek {self.__nama} dibuat")

    def get_nama(self):
        return self.__nama

    def set_nama(self, nama):
        self.__nama = nama

    def tampilkan(self):
        print(f"Nama: {self.__nama}")
        print(f"NIM: {self.__nim}")

    def __del__(self):
        print(f"Objek {self.__nama} dihapus")


if __name__ == "__main__":
    print("Program dimulai\n")

    mhs = Mahasiswa("Ulfan", "4.33.25.2.23")

    print("\nData awal:")
    mhs.tampilkan()

    mhs.set_nama("Ulfan Baru")

    print("\nSetelah diubah:")
    print(mhs.get_nama())

    del mhs

    print("\nProgram selesai")
```
